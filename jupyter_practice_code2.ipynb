```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

valorant = pd.read_csv("valorant-stats_PALOMALO_SILNO.csv")
```

```python
guns_prices = valorant["Price"]

guns_damage = valorant["HDMG_0"]

plt.plot(guns_prices, guns_damage, color="pink", marker="o", markersize=8, label="График соотношения цены и урона в тело")
plt.title("График соотношения цены и урона в тело")
plt.legend()
plt.xlabel("ЦЕНА")
plt.ylabel("УРОН")
plt.show()
```

```python
guns_damage = valorant["BDMG_0"]

plt.plot(guns_prices, guns_damage, color="blue", marker="o", markersize=8, label="График соотношения цены и урона в голову")
plt.title("График соотношения цены и урона в голову")
plt.legend()
plt.xlabel("УРОН")
plt.ylabel("ЦЕНА")
plt.show()
```

```python
guns_damage = valorant["LDMG_0"]

plt.plot(guns_prices, guns_damage, color="red", marker="o", markersize=8, label="График соотношения цены и урона в ноги")
plt.title("График соотношения цены и урона в ноги")
plt.legend()
plt.xlabel("ЦЕНА")
plt.ylabel("УРОН")
plt.show()
```

```python
guns_wall_penetration = valorant["Wall Penetration"]
guns_fire_rate = valorant["Fire Rate"]

plt.bar(guns_fire_rate, guns_wall_penetration, color="green", label="График соотношения скорости стрельбы и уровня прострела стен")
plt.title("График соотношения скорости стрельбы и уровня прострела стен")
plt.legend()
plt.xlabel("скорость стрельбы")
plt.ylabel("уровень прострела стен")
plt.show()
```

```python
guns_wall_penetration = valorant["Magazine Capacity"]
guns_fire_rate = valorant["Fire Rate"]

plt.bar(guns_fire_rate, guns_wall_penetration, color="purple", label="График соотношения скорости стрельбы и объёма магазина")
plt.title("График соотношения скорости стрельбы и объёма магазина")
plt.legend()
plt.xlabel("скорость стрельбы")
plt.ylabel("объём магазина")
plt.show()
```

```python
guns_wall_penetration = valorant["BDMG_0"]
guns_fire_rate = valorant["Fire Rate"]

plt.bar(guns_fire_rate, guns_wall_penetration, color="red", label="График соотношения скорости стрельбы и урона в тело")
plt.title("График соотношения скорости стрельбы и урона в тело")
plt.legend()
plt.xlabel("скорость стрельбы")
plt.ylabel("урон в тело")
plt.show()
```

```python
rifles_count = (valorant["Weapon Type"] == "Rifle").sum()
sidearm_count = (valorant["Weapon Type"] == "Sidearm").sum()
x = (rifles_count, sidearm_count)
plt.figure(figsize=(10,10))
plt.pie(x, labels=("Винтовки", "Пистолеты"), autopct="%1.1f%%")
plt.legend()
plt.title("Диаграмма соотношения количества винтовок и пистолетов")
plt.show()
```

```python
rifles_count = (valorant["Weapon Type"] == "SMG").sum()
sidearm_count = (valorant["Weapon Type"] == "Shotgun").sum()
x = (rifles_count, sidearm_count)
plt.figure(figsize=(10,10))
plt.pie(x, labels=("Пистолеты пулемёты", "Дробовики"), autopct="%1.1f%%")
plt.legend()
plt.title("Диаграмма соотношения количества пистолетов пулемётов и дробавиков")
plt.show()
```
